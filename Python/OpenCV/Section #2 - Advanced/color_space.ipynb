ðŸ” Summary of Each Color Space:
| Color Space | Use Case                                       | Description                                                        |
| ----------- | ---------------------------------------------- | ------------------------------------------------------------------ |
| `Grayscale` | Simpler processing (e.g., edge detection)      | Only light intensity, no color                                     |
| `HSV`       | Color-based segmentation (e.g., red detection) | Separates color (Hue) from brightness (Value)                      |
| `LAB`       | Advanced color analysis, perceptual difference | Human-like perception â€” used in object detection, color correction |


```python

import cv2 as cv

# Load the image
img = cv.imread('wal2.png')

# Convert to GRAYSCALE
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Convert to HSV (Hue, Saturation, Value)
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)

# Convert to LAB (Lightness, A, B)
lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)

# Show all results
cv.imshow("Original", img)
cv.imshow("Grayscale", gray)
cv.imshow("HSV", hsv)
cv.imshow("LAB", lab)

cv.waitKey(0)
cv.destroyAllWindows()
```
